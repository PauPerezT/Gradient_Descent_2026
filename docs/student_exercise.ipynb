## Equations

Model:
$$
\hat y = wx + b
$$

Loss:
$$
L(w,b) = \frac{1}{n}\sum_{i=1}^n \left((wx_i+b)-y_i\right)^2
$$

Gradients:
$$
\frac{\partial L}{\partial w} = \frac{2}{n}\sum_{i=1}^n x_i\left((wx_i+b)-y_i\right)
$$
$$
\frac{\partial L}{\partial b} = \frac{2}{n}\sum_{i=1}^n \left((wx_i+b)-y_i\right)
$$

Update:
$$
w^{(t+1)} = w^{(t)} - \eta \frac{\partial L}{\partial w}, \qquad
b^{(t+1)} = b^{(t)} - \eta \frac{\partial L}{\partial b}
$$

```python
# Helper code for running experiments

def prepare_feature(feature_name="bmi", scale_x=False):
    idx = feature_names.index(feature_name)
    x = X[:, idx].reshape(-1, 1)

    if scale_x:
        scaler = StandardScaler()
        x = scaler.fit_transform(x)

    return x.ravel(), y


def closed_form_solution(x, y):
    model = LinearRegression()
    model.fit(x.reshape(-1, 1), y)
    w = float(model.coef_[0])
    b = float(model.intercept_)
    mse = mean_squared_error(y, model.predict(x.reshape(-1, 1)))
    return w, b, mse


def run_experiment(feature_name="bmi", eta=0.1, epochs=200, scale_x=False):
    x, y_local = prepare_feature(feature_name, scale_x)
    w_gd, b_gd, history = gradient_descent(x, y_local, eta=eta, epochs=epochs)

    w_cf, b_cf, mse_cf = closed_form_solution(x, y_local)
    mse_gd = compute_loss(x, y_local, w_gd, b_gd)

    print(f"Feature: {feature_name} | scaled={scale_x}")
    print(f"Gradient Descent -> w={w_gd:.4f}, b={b_gd:.4f}, MSE={mse_gd:.4f}")
    print(f"Closed-form      -> w={w_cf:.4f}, b={b_cf:.4f}, MSE={mse_cf:.4f}")

    fig, axes = plt.subplots(1, 2, figsize=(12, 4))

    axes[0].scatter(x, y_local, alpha=0.5, label="data")
    xs = np.linspace(x.min(), x.max(), 200)
    axes[0].plot(xs, predict(xs, w_gd, b_gd), label="GD line")
    axes[0].plot(xs, predict(xs, w_cf, b_cf), label="Closed-form line", linestyle="--")
    axes[0].set_title("Fit")
    axes[0].set_xlabel(feature_name)
    axes[0].set_ylabel("target")
    axes[0].legend()

    axes[1].plot(history)
    axes[1].set_title("Loss during training")
    axes[1].set_xlabel("iteration")
    axes[1].set_ylabel("MSE")

    plt.tight_layout()
    plt.show()
```
